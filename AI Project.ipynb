```python
data['Disease'] = data['Disease'].fillna(method = 'ffill')
data['Count of Disease Occurrence'] = data['Count of Disease Occurrence'].fillna(method = 'ffill')

data
```

  <h1>--------------------------------------------------------------------------------------------------------</h1>
<div> <p> In step 3 I want to collect same Symptom in same Disease by sum the Count of Disease Occurrence for this Symptom so i want to use group by with sum to make it </p></div>


```python
data = data.groupby(['Disease','Symptom'])['Count of Disease Occurrence'].sum().reset_index()
data
```

<h1>--------------------------------------------------------------------------------------------------------</h1>
<div> <p> In step 5 I want to use Naive Bayes to account the Probability for each Symptom in each Disease , thats by sum the Count of Disease Occurrence for all Symptom in each Disease , then Dividing the Count of Disease Occurrence for each Symptom in the Total </p></div>

```python
data['Total'] = data['Count of Disease Occurrence'].groupby(data['Disease']).transform('sum')
data
```

```python
data['Probability'] = (data['Count of Disease Occurrence'] / data['Total'])
data
```
