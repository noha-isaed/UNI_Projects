```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df_player = pd.DataFrame(np.random.randn(1000, 2), columns=['page_views', 'market_value'])
df_player['market_value'] = df_player['market_value'] + np.arange(1000)
df_player.plot(kind='hexbin', x='page_views', y='market_value', gridsize=25)
```

```python
df_player.boxplot()
```

```python
sns.violinplot(data = df_player )
 
```

```python
sns.set_style("white")
sns.kdeplot(df_player.market_value, df_player.page_views, cmap="Blues" , shade=True ,shade_loest =True)
plt.show()
```
