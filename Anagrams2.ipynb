```python
wordclean =  [word.strip().lower() for word in open('words.txt','r')]
```

```python
wordclean[:10]
```

```python
wordclean = sorted(list(set([word.strip().lower()  for word in open('words.txt' , 'r')])))
```

```python
wordclean[:10]
```

```python
dictionary =  {}

for i in wordclean:
    
    simplified= ''.join(sorted(i)).lower()
    
    if simplified in dictionary :
        dictionary[simplified].append(i)
    else:
        dictionary[simplified]=[i]  
```

```python
grouplist = []
for i in dictionary:
    if len(dictionary[i]) > 1:
        grouplist.append(dictionary[i])
     
```

```python
finalDictionary = {}
for anagrams in grouplist:
    if len(anagrams[0]) in finalDictionary:
            finalDictionary[len(anagrams[0])].append(anagrams)
    else:
        finalDictionary[len(anagrams[0])] =[anagrams]
```
